```python
import pandas as pd

df = pd.read_csv('stock_data.csv')
print(df.head())
```

```python
data = pd.read_csv("stock_data.csv")

print(data.head())
print(data.info())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import Dense
```

```python
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
dataset = df['Stock_1'].values
dataset = dataset.reshape(-1,1)
```

```python
plt.figure(figsize=(10,5))

plt.plot(dataset)

plt.title("Stock_1 Price Trend")
plt.xlabel("Days")
plt.ylabel("Price")

plt.show()
```

```python
scaler = MinMaxScaler(feature_range=(0,1))

dataset_scaled = scaler.fit_transform(dataset)
```

```python
train_size = int(len(dataset_scaled)*0.8)

train_data = dataset_scaled[:train_size]
test_data = dataset_scaled[train_size:]
```

```python
def create_sequences(data, time_step):

    X = []
    y = []

    for i in range(len(data)-time_step):

        X.append(data[i:(i+time_step),0])

        y.append(data[i+time_step,0])

    return np.array(X), np.array(y)
```

```python
time_step = 10
```

```python
X_train, y_train = create_sequences(train_data,time_step)

X_test, y_test = create_sequences(test_data,time_step)
```

```python
print(X_train.shape)

print(y_train.shape)

print(X_test.shape)

print(y_test.shape)
```

```python
X_train = X_train.reshape(
    X_train.shape[0],
    X_train.shape[1],
    1
)

X_test = X_test.reshape(
    X_test.shape[0],
    X_test.shape[1],
    1
)
```

```python
print(X_train.shape)
```

```python
model = Sequential()

model.add(
    SimpleRNN(
        units=50,
        activation='tanh',
        input_shape=(time_step,1)
    )
)

model.add(Dense(1))
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test,y_test),
    epochs=50,
    batch_size=16,
    verbose=1
)
```

```python
plt.figure(figsize=(10,5))

plt.plot(history.history['loss'])

plt.plot(history.history['val_loss'])

plt.title("Training vs Validation Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.legend(['Training','Validation'])

plt.show()
```

```python
predictions = model.predict(X_test)
```

```python
predictions = scaler.inverse_transform(predictions)

actual_values = scaler.inverse_transform(
    y_test.reshape(-1,1)
)
```

```python
rmse = np.sqrt(
    mean_squared_error(
        actual_values,
        predictions
    )
)

print("RMSE =", rmse)
```

```python
mae = mean_absolute_error(
    actual_values,
    predictions
)

print("MAE =", mae)
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    actual_values,
    label='Actual'
)

plt.plot(
    predictions,
    label='Predicted'
)

plt.title("Actual vs Predicted Stock Price")

plt.xlabel("Time")

plt.ylabel("Stock Price")

plt.legend()

plt.show()
```
